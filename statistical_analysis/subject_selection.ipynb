```python
import os
import pandas as pd
```

```python
phenotypic_file = '../csv-fenotipico/Table_new_cols.csv'
phen_df = pd.read_csv(phenotypic_file)
display(phen_df)
```

```python
exp_dir = 'results_abr_22_2021'
group = 'results_children'
region = 'Right Frontal Medial Cortex'
```

```python
df = pd.DataFrame()
for p in os.walk(os.path.join(exp_dir, group, region, '1-sig')):
    for file in p[2]:
        if 'subjects' in file.split('-'):
            df = pd.concat([df, pd.read_csv(os.path.join(exp_dir, group, region, '1-sig', file))])
```

```python
pd.options.display.max_rows = 300
```

```python
ups_control = df.pivot_table(index=['2'], aggfunc='size').sort_values(ascending=False)
print(ups_control)
```

```python
ups_asd = df.pivot_table(index=['1'], aggfunc='size').sort_values(ascending=False)
print(ups_asd)
```

```python
control = pd.DataFrame(data={'SUB_ID': ups_control.index.values,
                            'subband-counts': ups_control.values}).merge(phen_df ,how='inner', on='SUB_ID')
control['DX_GROUP'] = 2

asd = pd.DataFrame(data={'SUB_ID': ups_asd.index.values,
                            'subband-counts': ups_asd.values}).merge(phen_df ,how='inner', on='SUB_ID')
asd['DX_GROUP'] = 1
```

```python
display(control.sort_values(by='subband-counts', ascending=False))
```

```python
display(asd.sort_values(by='subband-counts', ascending=False))
```
